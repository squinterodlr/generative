```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['agg.path.chunksize'] = 10000

%matplotlib inline
```

```python
def archimedean(theta,amplitude,root):
    
    return amplitude*(theta**(1/root))

def theta_archimedean(r,amplitude,root):
    
    return (r/amplitude)**root    
```

```python
rmax = 10
amplitude = 1
sqnplots = 4


roots = np.linspace(0.8,3,sqnplots*sqnplots)

fig, axs = plt.subplots(sqnplots,sqnplots)
fig.set_size_inches(10.80,10.80)

for i in range(sqnplots):
    for j in range(sqnplots):
        
        ax = axs[i,j]
        
        root = roots[i*sqnplots + j]

        Thetas = np.arange(0,theta_archimedean(rmax,amplitude,root),0.1)
        Rs = archimedean(Thetas,amplitude,root)
        
        ax.plot(Rs*np.cos(Thetas),Rs*np.sin(Thetas),'k',lw=1)
        ax.set_xlim(-1.05*rmax,1.05*rmax)
        ax.set_ylim(-1.05*rmax,1.05*rmax)
        ax.set_aspect('equal')
        ax.axis('off')
        
fig.patch.set_alpha(1)
fig.tight_layout()
fig.savefig('archimedean-spiral-grid.png',dpi=100)
plt.show()
```

```python
fps = 60
vidlength = 3


nturns = 5
offsets = np.linspace(0,nturns*2*np.pi,fps*vidlength,endpoint=False)


rmax = 10
amplitude = 1
sqnplots = 4
roots = np.linspace(0.8,3,sqnplots*sqnplots)
fig, axs = plt.subplots(sqnplots,sqnplots)
fig.set_size_inches(10.80,10.80)
fig.patch.set_alpha(1)
fig.tight_layout()

for frame,offset in enumerate(offsets):
    print('Frame ',frame+1,' of ',len(offsets),end='\r')
    for i in range(sqnplots):
        for j in range(sqnplots):

            ax = axs[i,j]
            ax.cla()
            
            root = roots[i*sqnplots + j]
            
            thisoffset = (2*np.pi)*(i*sqnplots + j)/(sqnplots*sqnplots)

            Thetas = np.arange(0,theta_archimedean(rmax,amplitude,root),0.1)
            Rs = archimedean(Thetas,amplitude,root)

            ax.plot(Rs*np.cos(Thetas-offset-thisoffset),Rs*np.sin(Thetas-offset-thisoffset),'k',lw=1)
            ax.set_xlim(-1.05*rmax,1.05*rmax)
            ax.set_ylim(-1.05*rmax,1.05*rmax)
            ax.set_aspect('equal')
            ax.axis('off')

    fig.savefig('animation/archimedean-spiral-grid%04d.png'%frame,dpi=100)
    
plt.show()
```
