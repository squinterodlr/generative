```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.integrate import solve_ivp
import perlin

plt.rcParams['agg.path.chunksize'] = 10000

rng = np.random.default_rng()
```

```python
perlin_factory = perlin.PerlinNoiseFactory(dimension=4,octaves=3)
```

```python
t = 0.9

Xs = np.linspace(0,2,100,endpoint=False)
Zs = np.array([[perlin_factory(x,y,np.cos(2*np.pi*t),np.sin(2*np.pi*t)) for y in Xs] for x in Xs])
print(Zs.min(), Zs.max())
plt.imshow(Zs)
plt.show()
```
