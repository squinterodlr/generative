```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import perlin
import palletes

%matplotlib inline

plt.rcParams['agg.path.chunksize'] = 10000

rng = np.random.default_rng()

count = 0
```

```python
n_turns = 5
n_lines = 3
background_color = 'k'
linewidth = 1
pallete = palletes.summer

Thetas = np.linspace(n_turns*2*np.pi,0,500,endpoint=False)
Theta_0s = np.linspace(0,2*np.pi,n_lines,endpoint=False)

Rs = (Thetas/(2*np.pi))**(1.5) #+ np.log(Thetas/(2*np.pi)) #+ np.cos(np.sqrt(2)*Thetas)

fig, ax = plt.subplots(figsize=(10.8,10.8))
fig.set_facecolor(background_color)



for theta_0 in Theta_0s:


    color = rng.choice(pallete)

    Xs = np.cos(Thetas-theta_0)
    Ys = np.sin(Thetas-theta_0)

    ax.plot(Rs*Xs, Rs*Ys, c=color, lw=linewidth)

    '''
    Xs = np.cos(-Thetas+theta_0)
    Ys = np.sin(-Thetas+theta_0)

    ax.plot(Rs*Xs, Rs*Ys, c=color, lw=linewidth)
    '''
ax.set_xlim(-1.1*Rs.max(),1.1*Rs.max())
ax.set_ylim(-1.1*Rs.max(),1.1*Rs.max())
ax.axis('off')
fig.tight_layout()
count = count+1
fig.savefig('images/spirals-%03d.png'%count,dpi=200,bbox_inches='tight',pad_inches=0)
plt.show()
```
