```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline

plt.rcParams['agg.path.chunksize'] = 10000

rng = np.random.default_rng()
```

## The basic tiles

```python
def truchet_tile(lower_left_pos,radius,**kwargs):
    
    upper_right = lower_left_pos + np.array([1-])
    
    bottom_arc = mpl.patches.Arc(lower_left_pos,2*radius,2*radius,theta1=0,theta2=90,**kwargs)
    upper_arc = mpl.patches.Arc(lower_left_pos+np.array([1,1]),2*radius,2*radius,theta1=180,theta2=270,**kwargs)
    
    return bottom_arc, upper_arc

def truchet_tile_reversed(lower_left_pos,radius,**kwargs):
    
    lower_right = lower_left_pos+np.array([1,0])
    upper_left = lower_left_pos+np.array([0,1])
    
    bottom_arc = mpl.patches.Arc(lower_right,2*radius,2*radius,theta1=90,theta2=180,**kwargs)
    upper_arc = mpl.patches.Arc(upper_left,2*radius,2*radius,angle= 270,theta1=0,theta2=90,**kwargs)
    return bottom_arc, upper_arc
```

```python
fig, axs = plt.subplots(1,2)

r_center = 0.5 #central radius

#bottom arc
bottom_arc, upper_arc = truchet_tile(np.array([0,0]),0.5)
axs[0].add_patch(bottom_arc)
axs[0].add_patch(upper_arc)

bottom_arc, upper_arc = truchet_tile_reversed(np.array([0,0]),0.5)
axs[1].add_patch(bottom_arc)
axs[1].add_patch(upper_arc)

plt.show()
```

```python
n_tiles = 20
radius = 0.5
Xs = np.linspace(0,n_tiles,n_tiles, endpoint=False)
#print(Xs)
fig,ax = plt.subplots(1)
fig.set_size_inches((10.8,10.8))
fig.patch.set_alpha(1)

tiles = [truchet_tile,truchet_tile_reversed]

for y in Xs:
    for x in Xs:
        tile = rng.choice(tiles)
        lower_left = np.array([x,y])
        #print(lower_left)
        
        bottom_arc, upper_arc = tile(lower_left,radius)
        ax.add_patch(bottom_arc)
        ax.add_patch(upper_arc)
        
ax.set_xlim(0,n_tiles)
ax.set_ylim(0,n_tiles)
ax.axis('off')
plt.show()
```
