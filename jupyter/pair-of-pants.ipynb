```python
def ease_cubic(t):

    if t < 0:
        return 0
    elif t < 0.5:
        return 4*t*t*t
    elif t < 1:
        return 1 - 0.5*(2-2*t)*(2-2*t)*(2-2*t)
    else:
        return 1

def ease_quintic(t):
   
    if t < 0:
        return 0
    elif t < 0.5:
        return 16*t*t*t*t*t
    elif t < 1:
        return 1 - 0.5*(2-2*t)*(2-2*t)*(2-2*t)*(2-2*t)*(2-2*t)
    else:
        return 1
```

```python
Ts = np.linspace(0,1,500)
Xs = np.array([ ease_quintic(2*t) if t < 0.5 else ease_quintic(1 - 2*(t-0.5)) for t in Ts])
#Xs = np.array([ease_quintic(t) for t in Ts])
plt.plot(Ts,Xs)
plt.show()
```

```python
n_frames = 60*5
background_color = 'k'


Ts = np.linspace(0,1,n_frames,endpoint=False)
Xs = np.linspace(-3,3,500)
Xs, Ys = np.meshgrid(Xs,Xs)

fig, ax = plt.subplots(figsize=(10.8,10.8))
fig.patch.set_alpha(1)
fig.set_facecolor(background_color)

for count, t in enumerate(Ts):

    if t < 0.5:
        x_shift = ease_quintic(2*t)
    else:
        x_shift = ease_quintic(1 - 2*(t-0.5) )

    ax.cla()

    Z1 = np.exp(-1*((Xs-2*x_shift)*(Xs-2*x_shift) + (Ys)*(Ys)))
    Z2 = np.exp(-1*((Xs+2*x_shift)*(Xs+2*x_shift) + (Ys)*(Ys)))
    ax.contour(Xs,Ys,Z1+Z2,levels=[np.exp(-1)],colors='w',linewidths=2)
    ax.set_xlim(-4,4)
    ax.set_ylim(-4,4)  
    ax.axis('off')
    fig.tight_layout()
    fig.savefig('animation/cell-division%04d.png'%count,dpi=200,bbox_inches='tight',pad_inches=0)
    print('Frame',count+1,'of',n_frames)
fig.show()
```
